```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
                "E:/Dokumentumok/adventofcode/2022/01/input.csv", 
                 skip_blank_lines=False, header = None, names=['calories'], 
#                 dtype = {'calories': pd.Int64Dtype()}
                )
```

```python
df.head(20)
```

```python
df['calories'] = df['calories'].fillna(999999999)
```

```python
z = 1

result = []
for value in df["calories"]:
    if value == 999999999:
        z += 1
    result.append(z)

result[:30]
```

```python
df['result'] = result
df.head(15)
```

```python
df = df[df.calories != 999999999]
```

```python
df.head(15)
```

```python
elfs_calories = df.groupby(['result']).sum()
```

```python
elfs_calories.head()
```

#### Question 1: How much calories is the elf carrying who carries the most calories?

```python
elfs_calories.max()
```

#### Question 2: How much calories are the 3 elves carrying who carry the most calories?

```python
elfs_calories.sort_values(by='calories', ascending = False)[:3].sum()
```
