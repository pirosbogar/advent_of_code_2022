```python
print(inventory.head())
print(len(inventory))
```

```python
round_counter = 1
inspection_counter = 1

new_inventory = pd.DataFrame(data, columns = ['round_cnt', 'monkey', 'worry_level'])

while round_counter <= 20:
    while inspection_counter <= len(inventory[inventory['round_cnt'] == round_counter]):
        for a, b, c in zip(inventory.round_cnt, inventory.monkey, inventory.worry_level):
            if b == 'monkey0':
                x = math.floor((c * 3) / 3)
                if x % 11 == 0:
                    inventory.append([round_counter + 1, 'monkey7', x])
                elif x % 11 != 0:
                    inventory.append([round_counter + 1, 'monkey2', x])

            elif b == 'monkey1':
                x = math.floor((c + 3) / 3)
                if x % 2 == 0:
                    inventory.append([round_counter + 1, 'monkey2', x])
                elif x % 2 != 0:
                    inventory.append([round_counter + 1, 'monkey0', x])

            elif b == 'monkey2':
                x = math.floor((c + 5) / 3)
                if x % 5 == 0:
                    inventory.append([round_counter + 1, 'monkey7', x])
                elif x % 5 != 0:
                    inventory.append([round_counter + 1, 'monkey5', x])

            elif b == 'monkey3':
                x = math.floor((c * 19) / 3)
                if x % 7 == 0:
                    inventory.append([round_counter + 1, 'monkey6', x])
                elif x % 7 != 0:
                    inventory.append([round_counter + 1, 'monkey4', x])

            elif b == 'monkey4':
                x = math.floor((c + 1) / 3)
                if x % 17 == 0:
                    inventory.append([round_counter + 1, 'monkey6', x])
                elif x % 17 != 0:
                    inventory.append([round_counter + 1, 'monkey1', x])

            elif b == 'monkey5':
                x = math.floor((c + 2) / 3)
                if x % 19 == 0:
                    inventory.append([round_counter + 1, 'monkey4', x])
                elif x % 19 != 0:
                    inventory.append([round_counter + 1, 'monkey3', x])

            elif b == 'monkey5':
                x = math.floor((c * c) / 3)
                if x % 3 == 0:
                    inventory.append([round_counter + 1, 'monkey0', x])
                elif x % 3 != 0:
                    inventory.append([round_counter + 1, 'monkey1', x])

            elif b == 'monkey6':
                x = math.floor((c + 8) / 3)
                if x % 13 == 0:
                    inventory.append([round_counter + 1, 'monkey3', x])
                elif x % 13 != 0:
                    inventory.append([round_counter + 1, 'monkey5', x])

            inspection_counter += 1

        round_counter += 1
   
    
```
