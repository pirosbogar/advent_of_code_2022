```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
                "E:/Dokumentumok/adventofcode/2022/02/input.txt", 
                 sep=' ',
                 skip_blank_lines=False, header = None, names=['opponent', 'me'])
```

```python
df.head(20)
```

```python
selection_score = []

for value in df['me']:
    if value == 'X':
        selection_score.append(1)
    elif value == 'Y':
        selection_score.append(2)
    else:
        selection_score.append(3)

print(selection_score[:20])
```

```python
outcome_score = []

for a, b in zip(df.opponent, df.me):
    if a == 'A' and b == 'X':
        outcome_score.append(3)
    elif a == 'B' and b == 'Y':
        outcome_score.append(3)
    elif a == 'C' and b == 'Z':
        outcome_score.append(3)
    elif a == 'A' and b == 'Y':
        outcome_score.append(6)
    elif a == 'A' and b == 'Z':
        outcome_score.append(0)
    elif a == 'B' and b == 'Z':
        outcome_score.append(6)
    elif a == 'C' and b == 'X':
        outcome_score.append(6)
    elif a == 'C' and b == 'Y':
        outcome_score.append(0)
    elif a == 'B' and b == 'X':
        outcome_score.append(0)

print(outcome_score[:20])
```

```python
df['selection_score'] = selection_score
df['outcome_score'] = outcome_score
df['total_score'] = df['selection_score'] + df['outcome_score']
df.head(20)
```

#### Question 1: What would your total score be if everything goes exactly according to your strategy guide (X, Y, Z meaning rock, paper, scissors)?

```python
df.total_score.sum()
```

#### Question 2: What would your total score be if everything goes exactly according to your strategy guide (X, Y, Z meaning lose, draw, win)?

```python
selected_shape = []

for a, b in zip(df.opponent, df.me):
    if a == 'A' and b == 'X':
        selected_shape.append('Z')
    elif a == 'B' and b == 'Y':
        selected_shape.append('Y')
    elif a == 'C' and b == 'Z':
        selected_shape.append('X')
    elif a == 'A' and b == 'Y':
        selected_shape.append('X')
    elif a == 'A' and b == 'Z':
        selected_shape.append('Y')
    elif a == 'B' and b == 'Z':
        selected_shape.append('Z')
    elif a == 'C' and b == 'X':
        selected_shape.append('Y')
    elif a == 'C' and b == 'Y':
        selected_shape.append('Z')
    elif a == 'B' and b == 'X':
        selected_shape.append('X')

print(selected_shape[:20])
```

```python
selection_score2 = []

for value in selected_shape:
    if value == 'X':
        selection_score2.append(1)
    elif value == 'Y':
        selection_score2.append(2)
    else:
        selection_score2.append(3)

print(selection_score2[:20])
```

```python
outcome_score2 = []

for a, b in zip(df.opponent, selected_shape):
    if a == 'A' and b == 'X':
        outcome_score2.append(3)
    elif a == 'B' and b == 'Y':
        outcome_score2.append(3)
    elif a == 'C' and b == 'Z':
        outcome_score2.append(3)
    elif a == 'A' and b == 'Y':
        outcome_score2.append(6)
    elif a == 'A' and b == 'Z':
        outcome_score2.append(0)
    elif a == 'B' and b == 'Z':
        outcome_score2.append(6)
    elif a == 'C' and b == 'X':
        outcome_score2.append(6)
    elif a == 'C' and b == 'Y':
        outcome_score2.append(0)
    elif a == 'B' and b == 'X':
        outcome_score2.append(0)

print(outcome_score2[:20])
```

```python
df['selection_score2'] = selection_score2
df['outcome_score2'] = outcome_score2
df['total_score2'] = df['selection_score2'] + df['outcome_score2']
df.head(20)
```
